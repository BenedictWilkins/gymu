```python
from pprint import pprint
from more_itertools import windowed

import jnu as J
import gymu
from gymu.data import Composable
import numpy as np

def window(window_size=2):
    def _window(source):
        for x in windowed(source, window_size):
            k = x[0].keys()
            x = [z.values() for z in x]
            yield dict(zip(k, [np.concatenate(z) for z in zip(*x)]))
    return _window

dataset = gymu.data.dataset("./episode.tar.gz")
dataset = dataset.then(window(3))
dataset = dataset.then(Composable.mode(gymu.mode.sa))

for x in dataset:
    print(x.state.shape, x.action.shape)
    
```
